```python
import pandas as pd
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import os

# For nice Plots
%matplotlib inline
plt.style.use('default')
# Get the directory or the images
im_dir = os.getcwd()+'\\Images\\'
```

```python
# reading the CSV file
df_pixels = pd.read_csv('HaitiPixels.csv')
 
# displaying the contents of the CSV file
print(df_pixels)
```

```python

class_names = np.unique(df_pixels['Class'])
plt.subplots(figsize=(12,7))
for name in class_names:
  df_pixels_thisclass = df_pixels[df_pixels['Class']==name]
  plt.scatter(df_pixels_thisclass['Red'], df_pixels_thisclass['Blue'], label=name, s=5, alpha=0.5)
plt.xlabel('Red Pixel Intensity')
plt.ylabel('Blue Pixel Intensity')
plt.title('Scatterplot of Haiti Pixel Data, Colored by Class Labels');
plt.legend();
```

```python
# reading the CSV file
df_holdout = pd.read_csv('hold_out_data.csv')
 
# displaying the contents of the CSV file
print(df_holdout)

class_names = np.unique(df_holdout['Class'])
plt.subplots(figsize=(12,7))
for name in class_names:
  df_holdout_thisclass = df_holdout[df_holdout['Class']==name]
  plt.scatter(df_holdout_thisclass['Red'], df_holdout_thisclass['Blue'], label=name, s=5, alpha=0.5)
plt.xlabel('Red Pixel Intensity')
plt.ylabel('Blue Pixel Intensity')
plt.title('Scatterplot of Holdout Haiti Pixel Data, Colored by Class Labels');
plt.legend();
```
